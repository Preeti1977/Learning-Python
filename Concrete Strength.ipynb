```python
#TRAINING AND TESTING DATA SPLIT
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,shuffle=True)
train_test_split(y, random_state=46)
y_test
```

```python
#RANDOM FOREST FOR CONCRETE STRENGTH PREDICTION

from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor

from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor(n_estimators=45,random_state=50,max_depth=5, bootstrap=True, criterion='mse',
                       max_features='auto',max_leaf_nodes=None,
                      max_samples=None, min_impurity_decrease=0.01,
                      min_impurity_split=None, min_samples_leaf=15,
                      min_samples_split=5, min_weight_fraction_leaf=0.001,
                      n_jobs=None, oob_score=True,
                     verbose=0, warm_start=False)
model=regressor.fit(X_train,y_train.values.ravel())
y_pred=model.predict(X_test)
y_pred       
```

```python
#RANDOM FOREST ERROR

from sklearn import metrics 
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import numpy as np
print('Mean Absolute  error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared error:',metrics.mean_squared_error(y_test,y_pred))
print('Root Mean square error:',np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
print('coefficient of determination:', metrics.r2_score(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt
# plotting the data
plt.scatter(y_test, y_pred)
plt.title("Concrete Strength-Random Forest")
plt.xlabel('Observed Strength')
plt.ylabel('Predicted Strength')
plt.show()
```

```python
#ERROR FOR LINEAR REGRESSION
from sklearn import metrics 
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import numpy as np
print('Mean Absolute  error:', metrics.mean_absolute_error(y_test, y_pred1))
print('Mean Squared error:',metrics.mean_squared_error(y_test,y_pred1))
print('Root Mean square error:',np.sqrt(metrics.mean_squared_error(y_test,y_pred1)))
print('coefficient of determination:', metrics.r2_score(y_test,y_pred1))
```
